Data Split


```python
X = credit_card_data.drop (columns ='Class', axis = 1)
Y = credit_card_data['Class']
```

```python
print(X)
```

```python
print (Y)
```

Split Data into train and test set

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size =0.2, stratify =Y, random_state = 2 )
```

```python
print (X.shape,X_train.shape, X_test.shape)
```

Model Training

Logistic Regression


```python
model = LogisticRegression()
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


model = LogisticRegression(max_iter=2000)
model.fit(X_train_scaled, Y_train)
```

Model Evaulation

Accuracy Score

```python
x_train_prediction = model.predict(X_train_scaled)
training_data_accuarcy = accuracy_score(x_train_prediction, Y_train)
```

```python
print('Accuarcy on Training:', training_data_accuarcy)
```

```python
x_test_prediction = model.predict(X_test_scaled)
test_data_accuarcy = accuracy_score(x_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data: ',test_data_accuarcy)
```
